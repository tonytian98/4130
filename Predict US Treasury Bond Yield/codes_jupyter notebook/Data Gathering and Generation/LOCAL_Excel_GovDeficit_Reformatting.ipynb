```python
import pandas as pd
def read(filename,skip=10,index=0):
    try:
        xls = pd.ExcelFile(r'C:\Users\Admin\Downloads\{}.xls'.format(filename))

        df = xls.parse( skiprows=skip, index_col=index)
        psave(df,filename)
        return df
    except:
        print('{} failed'.format(filename))

### save and load data using pickle
import pickle
def psave(dataframe, fname):
    with open('{}.pkl'.format(fname),'wb') as pickle_file:
        pickle.dump(dataframe,pickle_file)
        
def pload(fname):
    with open('{}.pkl'.format(fname),'rb') as pickle_file:
        dataframe=pickle.load(pickle_file)
        return dataframe
```

```python
dt=pd.read_csv(r'C:\Users\Admin\Downloads\def2.csv')
print(dt)
```

```python
###change the above data to a nicer format
loc=pd.unique(dt['LOCATION'])


def add(x):
    return x+'-01-01'

def add_country(x):
    return x+'deficit'

countries=['USA' , 'BRA', 'CHE' , 'CHN' ,'GBR' ,'IRL' , 'JPN' ,'LUX']
columns=[add_country(x) for x in countries ]
def dat(country):
    df1 = dt[dt['LOCATION']  ==country]
    df2=df1.drop(columns=['LOCATION','INDICATOR','SUBJECT','MEASURE','Flag Codes','FREQUENCY'])
    ddf2=df2.set_index('TIME')
    indexrn={}
    for i in ddf2.index:
        x=str(i)    
        indexrn[i]=add(x)
        ddf2=ddf2.rename(index=indexrn)
    clrn={}
    for i in ddf2.columns:
        clrn[i]=add_country(country)
        ddf2=ddf2.rename(columns=clrn)
    return ddf2
```

```python
d={}
for c in countries:
    d[c]=dat(c)
```
