```python
from  google.colab import drive  
drive.mount('/content/drive')
%cd /content/drive/My\ Drive/econ4130
### save and load data using pickle
import pickle
def psave(dataframe, fname):
    with open('{}.pkl'.format(fname),'wb') as pickle_file:
        pickle.dump(dataframe,pickle_file)
        
def pload(fname):
  
  with open('{}.pkl'.format(fname),'rb') as pickle_file:
      dataframe=pickle.load(pickle_file)
      return dataframe

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DF=pload('11_6')
```

```python
DF
```

```python
###Generate technical indicators
od = DF.filter(['1Y','2Y','3Y','5Y','7Y','10Y','20Y','30Y'],axis=1)
print(od)

df_SMA_5= od.rolling(5).sum()/5
df_SMA_5=df_SMA_5.add_suffix('_SMA5')
```

```python
df_SMA_5
```

```python
###moving averages
#10-day MA
df_SMA_10_0= od.rolling(10).sum()/10
df_SMA_10=df_SMA_10_0.add_suffix('_SMA10')
#15-day MA
df_SMA_15= od.rolling(15).sum()/15
df_SMA_15=df_SMA_15.add_suffix('_SMA15')
#20-day MA
df_SMA_20= od.rolling(20).sum()/20
df_SMA_20=df_SMA_20.add_suffix('_SMA20')
#50-day MA
df_SMA_50= od.rolling(50).sum()/50
df_SMA_50=df_SMA_50.add_suffix('_SMA50')
#200-day MA
df_SMA_200= od.rolling(200).sum()/200
df_SMA_200=df_SMA_200.add_suffix('_SMA200')

df_SMA_4= od.rolling(4).sum()/4
df_SMA_14= od.rolling(14).sum()/14
df_SMA_24= od.rolling(24).sum()/24
df_SMA_35= od.rolling(35).sum()/35
df_SMA_48= od.rolling(48).sum()/48
```

```python
###moving average spreads
# 10-day MA minus 4-day MA
df_SMA10_4=df_SMA_10_0.subtract(df_SMA_4)
df_SMA10_4=df_SMA10_4.add_suffix('_SMA10_4')
df_SMA10_4.info()
# 24-day MA minus 14-day MA
df_SMA24_14=df_SMA_24.subtract(df_SMA_14)
df_SMA24_14=df_SMA24_14.add_suffix('_SMA24_14')
# 48-day MA minus 35-day MA
df_SMA48_35=df_SMA_48.subtract(df_SMA_35)
df_SMA48_35=df_SMA48_35.add_suffix('_SMA48_35')
```

```python
df_SMA = pd.concat([df_SMA_5,df_SMA_10,df_SMA_15,df_SMA_20,df_SMA_50,df_SMA_200,
                    df_SMA10_4,df_SMA24_14,df_SMA48_35],axis=1,sort=False)
psave(df_SMA,'SMA1_1')
```

```python
### Generate bond spread data
d=pload('11_6')
d['JGB1Y']
```

```python
# US JAPAN BOND SPREAD
bdsp=pd.DataFrame()
for i in ['1Y','2Y','3Y','5Y','7Y','10Y','20Y','30Y']:
  name='USJPBDSP'+i
  jp='JGB'+i
  bdsp[name]=d[i]-d[jp]
```

```python
bdsp
```

```python
#US EU BOND SPREAD
for i in ['2Y','5Y','7Y','10Y','3Y']:
  name='USEUBDSP'+i
  jp='ECB'+i
  bdsp[name]=d[i]-d[jp]
```

```python
bdsp
```

```python
###US intra-market (Different time to maturity) bond spread
l=['1Y','2Y','3Y','5Y','7Y','10Y','20Y','30Y']
for i in range(len(l)):
  ii=l[i]
  a=l[i+1:]
  for j in a:
    name= 'US'+ii+'-'+j
    bdsp[name]=d[ii]-d[j]
bdsp
```

```python
##save bond spread data
psave(bdsp,'bdsp')
```
